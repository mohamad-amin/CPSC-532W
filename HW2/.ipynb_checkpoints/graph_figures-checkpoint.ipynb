```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from graph_based_sampling import *

%matplotlib inline
```

```python
def plot_hist(vals, figsize=(10, 5)):
    plt.figure(figsize=figsize)
    plt.hist(vals, bins=np.histogram(vals)[1])
    plt.show()
    
def plot_2_hists(vals):
    plt.subplot(1, 2, 1)
    plt.hist(vals[:, 0], bins=np.histogram(vals[:, 0])[1])
    plt.subplot(1, 2, 2)
    plt.hist(vals[:, 1], bins=np.histogram(vals[:, 1])[1])
    plt.tight_layout()
    plt.show()
    
```

# Program 1

```python
graph = pkl.load(open('programs/1_graph.pkl', 'rb'))
result = []
for i in range(10000):
    result.append(sample_from_joint(graph))
result = torch.stack(result).numpy()
plot_hist(result, figsize=(8, 6))
```

# Program 2

```python
graph = pkl.load(open('programs/2_graph.pkl', 'rb'))
result = []
for i in range(10000):
    result.append(sample_from_joint(graph))
result = torch.stack(result).numpy()

plt.figure(figsize=(10, 6))
plot_2_hists(result)
```

# Program 3

```python
graph = pkl.load(open('programs/3_graph.pkl', 'rb'))
result = []
for i in range(10000):
    result.append(sample_from_joint(graph))
result = torch.stack(result).numpy()

plt.figure(figsize=(15, 9))
for i in range(3):
    for j in range(6):
        ind = i*6 + j
        if ind >= 17:
            break
        plt.subplot(3, 6, ind+1)
        plt.hist(result[:, ind], bins=np.histogram(result[:, ind])[1])
plt.tight_layout()
plt.show()
```

# Program 4

```python
graph = pkl.load(open('programs/4_graph.pkl', 'rb'))
w0s, b0s, w1s, b1s = [], [], [], []
for i in range(1000):
    res = sample_from_joint(graph)
    w0s.append(res[0])
    b0s.append(res[1])
    w1s.append(res[2])
    b1s.append(res[3])
```

```python
w0s = torch.stack(w0s).numpy().squeeze()
plt.figure(figsize=(15, 2.5))
plt.subplot(2, 1, 1)
sns.heatmap(np.atleast_2d(w0s.mean(axis=0)), annot=True)
plt.title('Means')
plt.subplot(2, 1, 2)
sns.heatmap(np.atleast_2d(w0s.std(axis=0)), annot=True)
plt.title('STDs')
plt.tight_layout()
plt.show()
```

```python
b0s = torch.stack(b0s).numpy().squeeze()
plt.figure(figsize=(15, 2.5))
plt.subplot(2, 1, 1)
sns.heatmap(np.atleast_2d(b0s.mean(axis=0)), annot=True)
plt.title('Means')
plt.subplot(2, 1, 2)
sns.heatmap(np.atleast_2d(b0s.std(axis=0)), annot=True)
plt.title('STDs')
plt.tight_layout()
plt.show()
```

```python
w1s = torch.stack(w1s).numpy().squeeze()
plt.figure(figsize=(17, 8))
plt.subplot(1, 2, 1)
sns.heatmap(np.atleast_2d(w1s.mean(axis=0)), annot=True)
plt.title('Means')
plt.subplot(1, 2, 2)
sns.heatmap(np.atleast_2d(w1s.std(axis=0)), annot=True)
plt.title('STDs')
plt.tight_layout()
plt.show()
```

```python
b1s = torch.stack(b1s).numpy().squeeze()
plt.figure(figsize=(15, 2.5))
plt.subplot(2, 1, 1)
sns.heatmap(np.atleast_2d(b1s.mean(axis=0)), annot=True)
plt.title('Means')
plt.subplot(2, 1, 2)
sns.heatmap(np.atleast_2d(b1s.std(axis=0)), annot=True)
plt.title('STDs')
plt.tight_layout()
plt.show()
```
